## Introdução

A minha principal motivação era descobrir como é a distribuição salarial para diferentes aposentados, diferenciando-os principalmente por gênero e cor da pele, e se tinha mudado desde 2001.

Essa pesqusa diz que o salários de negros são quase 40% menores do que o de brancos.

Eu queria descobrir se esses dados se tornam verdade quando estudados em um outro grupo, o dos aposentados.
